## ***데이터 세트 분리***

---








```python
import matplotlib.pyplot as plt # 데이터 시각화용
import pandas as pd # 데이터 가공하는 용
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/PythonMLWorkspace/SciKitLearn/LinearRegressionData.csv')
dataset
```

```python
X = dataset.iloc[:, :-1].values # 마지막 컬럼은 남겨둔다
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split # 훈련 셋과 테스트 셋을 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) # 훈련 80, 테스트 20으로 분리
```

```python
X, len(X) # 전체 데이터 X, 개수
```

```python
X_train, len(X_train) # 훈련 데이터 X, 개수
```

```python
X_test, len(X_test) # 테스트 데이터 X, 개수
```

```python
y, len(y) # 전체 데이터 y
```

```python
y_train, len(y_train) # 훈련세트 y
```

```python
y_test, len(y_test) # 테스트 세트 y
```

### 분리된 데이터를 통한 모델링

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train) # 훈련 세트로 학습
```

### 데이터 시각화 (훈련세트)

```python
plt.scatter(X_train, y_train, color='red') # 산점도 그래프
plt.plot(X_train, reg.predict(X_train), color='green') # 선 그래프
plt.title('Score by hours (train data)') # 제목
plt.xlabel('hours') # x축 이름
plt.ylabel('score') # y축 이름
plt.show() # 화면에 보여주기
```

### 데이터 시각화 (테스트 세트)

```python
plt.scatter(X_test, y_test, color='red') # 산점도 그래프 (테스트 만 점으로 찍어서)
plt.plot(X_train, reg.predict(X_train), color='green') # 선 그래프 (학습 선은 그대로 둬야 함)
plt.title('Score by hours (test data)') # 제목
plt.xlabel('hours') # x축 이름
plt.ylabel('score') # y축 이름
plt.show() # 화면에 보여주기
```

```python
reg.coef_
```

```python
reg.intercept_
```

### 모델 평가

```python
reg.score(X_test, y_test) # 테스트 세트를 통한 모델 평가
```

```python
reg.score(X_train, y_train) # 훈련 세트를 통한 모델 평가
```

## 경사 하강법 (Gradient Descent)

max_iter : 훈련 세트 반복 횟수 (Epoch 횟수)

eta0 : 학습률 (learning rate)

```python
from sklearn.linear_model import SGDRegressor # SGD : Stochastic Gradient Descent 확률적 경사 하강법

#지수 표기법
# 1e-3 : 0.001 (10^-3)

sr = SGDRegressor(max_iter=1000, eta0=1e-4, random_state=0, verbose=1)
sr.fit(X_train, y_train)
```

```python
plt.scatter(X_train, y_train, color='red') # 산점도 그래프
plt.plot(X_train, sr.predict(X_train), color='green') # 선 그래프
plt.title('Score by hours (train data, SGD)') # 제목
plt.xlabel('hours') # x축 이름
plt.ylabel('score') # y축 이름
plt.show() # 화면에 보여주기
```

```python
sr.coef_, sr.intercept_
```

```python
sr.score(X_test, y_test) # 테스트 세트를 통한 모델 평가
```

```python
sr.score(X_train, y_train) # 훈련 세트를 통한 모델 평가
```
